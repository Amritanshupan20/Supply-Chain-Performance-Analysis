```python
df.drop(columns=['Product Description'], inplace=True)
```

```python
df.drop(columns=['Order Zipcode'], inplace=True)
```

```python
df['Customer Lname'] = df['Customer Lname'].fillna("Unknown")
```

```python
df['Customer Zipcode'] = df['Customer Zipcode'].fillna(df['Customer Zipcode'].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
print("Total Sales: $", round(df['Sales'].sum(), 2))
```

```python
print("Total Profit: $", round(df['Order Profit Per Order'].sum(), 2))
```

```python
print("Average Delivery Time:", round(df['Days for shipping (real)'].mean(), 2), "days")
```

```python
print("Total Customers:", df['Customer Id'].nunique())
```

```python
top_categories = (
    df.groupby('Category Name')['Sales']
      .sum()
      .sort_values(ascending=False)
)

top_categories.head(10)
```

```python
plt.figure(figsize=(10,6))

top_categories.head(10).plot(kind='bar')

plt.title("Top 10 Categories by Sales")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
top_products = (
    df.groupby('Product Name')['Sales']
      .sum()
      .sort_values(ascending=False)
)

top_products.head(10)
```

```python
plt.figure(figsize=(12,6))

top_products.head(10).plot(kind='bar')

plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.xticks(rotation=90)

plt.tight_layout()

plt.show()
```

```python
delivery_status = df['Delivery Status'].value_counts()

delivery_status
```

```python
plt.figure(figsize=(8,5))

delivery_status.plot(kind='bar')

plt.title("Delivery Status")

plt.xlabel("Status")
plt.ylabel("Orders")

plt.tight_layout()

plt.show()
```

```python
shipping_mode = df['Shipping Mode'].value_counts()

shipping_mode
```

```python
plt.figure(figsize=(8,5))

shipping_mode.plot(kind='bar')

plt.title("Shipping Mode")

plt.xlabel("Mode")
plt.ylabel("Orders")

plt.tight_layout()

plt.show()
```

```python
top_country = (
    df.groupby('Order Country')['Sales']
      .sum()
      .sort_values(ascending=False)
)

top_country.head(10)
```

```python
plt.figure(figsize=(10,6))

top_country.head(10).plot(kind='bar')

plt.title("Top 10 Countries by Sales")

plt.xlabel("Country")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
market_sales = (
    df.groupby('Market')['Sales']
      .sum()
      .sort_values(ascending=False)
)

market_sales
```

```python
plt.figure(figsize=(10,6))

market_sales.plot(kind='bar')

plt.title("Sales by Market")

plt.xlabel("Market")
plt.ylabel("Sales")

plt.tight_layout()

plt.show()
```

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])

monthly_sales = (
    df.groupby(df['order date (DateOrders)'].dt.to_period('M'))['Sales']
      .sum()
)

monthly_sales.index = monthly_sales.index.astype(str)
```

```python
plt.figure(figsize=(15,6))

plt.plot(monthly_sales.index, monthly_sales.values)

plt.title("Monthly Sales Trend")

plt.xlabel("Month")
plt.ylabel("Sales")

plt.xticks(rotation=90)

plt.tight_layout()

plt.show()
```

```python
df.to_csv("clean_supply_chain.csv", index=False)

print("Dataset Saved Successfully")
```
